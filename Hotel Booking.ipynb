# Preprocessing

```python
cat_feature=df.columns[df.dtypes==object]
cat_feature
```

```python
df[cat_feature].describe()
```

skema preprocessing:  

one hot encoding : hotel, distribution_channel, market_segment, customer_type, reservation_status   
ordinal : reserved_room_type   
  
no treatment : numerical

```python
onehot=OneHotEncoder()
onehot.fit_transform(df)
```

```python
#ordinal
ordinal_mapping = [
    {'col':'reserved_room_type',
    'mapping':{None:0,'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8,'L':9,'P':10}}
    ]
```

```python
ordinal_encoder = ce.OrdinalEncoder(mapping = ordinal_mapping)
ordinal_encoder
```

```python
transformer = ColumnTransformer([
    ('one hot encoder',OneHotEncoder(drop = 'first'),['hotel','distribution_channel','market_segment','customer_type','reservation_status']),
    ('ordinal encoder',ordinal_encoder,['reserved_room_type'])
], remainder = 'passthrough')
```

```python
transformer.fit_transform(df)
```

```python
df.head()
```
